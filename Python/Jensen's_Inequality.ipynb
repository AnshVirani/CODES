## Expectation of a Discrete Random Variable

In probability theory and statistics, the expectation, also known as the expected value, is a key concept that represents the average value of a random variable.

### Definition

For a discrete random variable $X$, which takes on a finite or countably infinite number of possible values $x_1, x_2, x_3, \ldots$, each with a probability $P(X = x_i)$, the expectation of $X$, denoted as $E[X]$ or $\mu$, is calculated as the weighted sum of the possible values of $X$, where each value is weighted by its corresponding probability:

$$ E[X] = \sum_{i} x_i \cdot P(X = x_i) $$

### Interpretation

The expectation of a random variable represents the average value that one would expect to obtain if the experiment or process is repeated a large number of times. It is a measure of the "center" or "balance point" of the probability distribution of $X$.

### Properties

1. **Linearity of Expectation**: If $X$ and $Y$ are random variables, and $a$ and $b$ are constants, then:
   $$ E[aX + bY] = aE[X] + bE[Y] $$
   
2. **Expectation of a Function of a Random Variable**: If $g(X)$ is a function of the random variable $X$, then:
   $$ E[g(X)] = \sum_{i} g(x_i) \cdot P(X = x_i) $$

### Importance

The expectation is a fundamental concept in probability and statistics and is used extensively in various applications, including decision theory, economics, and machine learning. It provides a single numerical summary of the behavior of a random variable and helps in making predictions and analyzing uncertainty.

```python
def expectation(X):
    """
    Computes the expectation of a discrete random variable X given its probability mass function (PMF).

    Args:
    - X (list of tuples): Probability mass function (PMF) of X represented as a list of tuples in the format (value, probability).

    Returns:
    - E_X (float): Expectation of X, calculated as the sum of each value weighted by its corresponding probability.
    """
    return sum(value * probability for value, probability in X)
```

## Jensen's Inequality

Jensen's Inequality is a fundamental concept in mathematics and probability theory. It states that for a convex function $f(x)$, the expectation of $f(X)$ is greater than or equal to $f$ applied to the expectation of $X$, i.e.,

$$ E[f(X)] \geq f(E[X]) $$

where $X$ is a random variable.


```python
# Define a function to test Jensen's inequality
def jensens_inequality_test(func, X):
    """
    Tests Jensen's inequality for a given function and random variable X.

    Args:
    - func (function): The function for which Jensen's inequality is to be tested.
    - X (list of tuples): Probability mass function (PMF) in the format (value, probability).

    Returns:
    - lhs (float): Left-hand side of Jensen's inequality.
    - rhs (float): Right-hand side of Jensen's inequality.
    """
    # Compute the expectation of the function
    E_fx = expectation([(func(value), probability) for value, probability in X])

    # Compute the function applied to the expectation of X
    fx_E_X = func(expectation(X))

    return E_fx, fx_E_X
```

## Testing

```python
# Generate 5 different X arrays
X_arrays = [
    [(0, 0.2), (1, 0.3), (2, 0.5)],
    [(1, 0.1), (2, 0.2), (3, 0.3), (4, 0.4)],
    [(0, 0.25), (1, 0.25), (2, 0.25), (3, 0.25)],
    [(0, 0.4), (5, 0.3), (10, 0.2), (15, 0.1)],
    [(0, 0.5), (1, 0.25), (2, 0.15), (3, 0.1)]
]

# Functions to test Jensen's inequality
functions = [square_func, identity_func, exponential_func, logarithmic_func]
```

```python
# Test Jensen's inequality for each combination of X and function
for X in X_arrays:
    for func in functions:
        print("Current X:", X)
        print("Testing function:", func.__name__,'\n')
        E_fx, fx_E_X = jensens_inequality_test(func, X)
        print("Expectation of f(X):", E_fx)
        print("f(Expectation of X):", fx_E_X, '\n')
        if E_fx >= fx_E_X:
            print("Jensen's inequality holds.\n\n", "="*60, '\n')
        else:
            print("Jensen's inequality does not hold.\n\n", "="*60, '\n')
```
